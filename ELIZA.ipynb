```python
class userProfile:
    def __init__(self, name):
        self.name = name
        self.age = 0
        self.rememberer = False
        self.dreamWordsUsed = []
        self.dreamThemesUsed = []                # example dreamThemesUsed ["animals", "vehicles", "animals"]
        self.dreamSentencesUsed = []
        self.sentenceLog = []
        self.themesLog = {}                      # example themesLog {"animals": 2, "vehicles": 1}
        self.themesResponded = []         # So I don't tell user same thing
    
    # Determine the number of times a dream theme has been used
    def dreamTheme(self):
        self.themesLog.clear()                               # Clears the log each time before adding them up.
        for key in dreamWordsDict:                           # Counts the number of times a theme has been used,
            themeCount = self.dreamThemesUsed.count(key)     # using the dreamThemesUsed log that is continually added to
            if themeCount > 0:                               # As long as it has been used at least once.
                self.themesLog[key] = themeCount             
    
    # Deterimine next highest dream theme usage
    def dreamThemeCounter(self):
        themeLargest = 0
        for key in self.themesLog:                   # Checks to see which theme was used the most
            if self.themesLog[key] > themeLargest:
                keyCheck = key
                #print(keyCheck)
                themeLargest = self.themesLog[key]
                #print(themeLargest)
        return(keyCheck)                               # Returns the currently most used theme
            
    # Keep track of what dream themes the user has already been told about
    def dreamThemesResponded(self, key):
        self.themesResponded.append(key)     #if test > 0 means the theme was already logged once
        
        
```

```python
# functions all in one place

def strip_punctuation(sentence):
    puncMark = ",.?;!"
    for char in puncMark:
        sentence = sentence.replace(char, " ")
    sentence = sentence.replace("  ", " ")
    sentence = sentence.replace("   ", " ")
    sentence = sentence.lower()
    sentence = sentence.strip()
    return sentence

def reflect_sentence(sentence, pronouns):  # Could have done 'global pronounDict' and not passed pronouns
    wordList = sentence.split()            # Turns the sentence into a list of seperate words
    for i in range(0, len(wordList)):
        if pronouns.get(wordList[i]) != None:
            wordList[i] = pronouns.get(wordList[i])
    newSentence = "Why do you think " + " ".join(wordList) + "?\n"
    return newSentence

def check_contractions(sentence):
    global contractionsDict
    wordList = sentence.split()                    # Turn the sentence into a list of seperate words
    for i in range (0, len(wordList)):
        if contractionsDict.get(wordList[i]) != None:       # Find and fix contractions
            wordList[i] = contractionsDict.get(wordList[i])
    newSentence = " ".join(wordList)
    return newSentence

def check_complex_sentence(sentence):
    checkSentence = sentence[0:-1] # Strip off the last (possibly) punctuation mark so I can look for 
    puncMark = ".?!"               # other punctuation, indicating two sentences
    test = False
    for char in puncMark:
        if char in checkSentence:
            test = True
    if test == True:
        return "complexsentence"     # If two or more sentences found return key word 'complexsentence'
    else:
        return sentence              # Otherwise return original sentence

# Asks for a name if input is first name last name it returns first name capitalized and comment
def get_name(sentence):
    nameList = sentence.split()     #turn the sentence into a list of seperate words
    if len(nameList) > 1:
        print("We should just stick with first names!\n")
    firstName = nameList[0]
    firstName = firstName.capitalize()
    return firstName

# This function checks to see if the sentence is about a dream if so it checks each word in the dream
# dictionary and logs it into the profile.
def check_dream_words(sentence):
    wordList = sentence.split()            #turn the sentence into a list of seperate words
    for i in range (0, len(wordList)):
        if wordList[i] in dreamWords:                      #check for dreaming words in sentence
            currentUser.dreamSentencesUsed.append(sentence)  # log the dream word
            sentence = "dream"                        #key word flag found dream sentence
            for j in range (0,len(wordList)):         #check every word in the sentence for a dream word
                for key in dreamWordsDict:
                    if wordList[j] in dreamWordsDict[key]:
                        currentUser.dreamWordsUsed.append(wordList[j])
                        currentUser.dreamThemesUsed.append(key)
                    

            
    return(sentence)                                 #return key work dream or original sentence


# This is the function that determines a dream theme and response
def create_dream_response():
    if len(currentUser.dreamThemesUsed) < 4:                            # If only 1, 2 or 3 dreams get another one
        return(random.choice(exclamatoryWords) + ", do you have a couple more examples?")
    currentUser.dreamTheme()                                            # Caluclate dream theme usage
    # *** need to check if there is more than one largest theme?***
    currentTheme = currentUser.dreamThemeCounter()
    #print(currentTheme)
    number = currentUser.themesLog[currentTheme]
    #print(number)
    #print(currentUser.themesResponded)
    if currentTheme in currentUser.themesResponded:           # if true we already told user about this theme
        return("\nOh", currentTheme, " again, we already talked about that.")
    else:
        currentUser.themesResponded.append(currentTheme)              # Add to those that have been output to user
        returnSentence = "\nDreaming about " + currentTheme + " " + random.choice(themeStarters) + " " + dreamThemeMeanings[currentTheme] + "\n"
        return(returnSentence)     # The money line. Return something about the dreams

```
